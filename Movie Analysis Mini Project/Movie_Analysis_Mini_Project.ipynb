#기초 세팅 및 데이터 분석

```
import pandas as pd
import seaborn as sns

#csv 파일 읽어오기
df_credits = pd.read_csv("/content/Movie Analysis/tmdb_5000_credits.csv")
df_movies = pd.read_csv("/content/Movie Analysis/tmdb_5000_movies.csv")
```

```
#데이터 내용 미리보기
#출력해본 결과, 단순 컬럼이 아니라 dictionary 형태로 되어있는 컬럼도 존재(object)
print(df_credits.head())
print('\n')
print(df_movies.head())
```

```
#데이터프레임의 기본 정보 출력
print(df_credits.info())
print('\n')
print(df_movies.info())
```

```
#데이터 사전 처리 시작!(시도라도)

#누락 데이터 확인
#NaN 개수 확인
nan_credits = df_credits['crew'].value_counts(dropna=False)
print(nan_credits)
print('\n')

#누락 데이터 직접 찾기
##영화정보 csv파일에 NaN값이 있음 -> 누락개수가 제각각이라 해당 열들을 삭제하기에는...
print(df_credits.isnull())
print("영화 크레딧파일 NaN 개수 : \n",df_credits.isnull().sum(axis=0))
print(df_movies.isnull())
print("영화정보파일 NaN 개수 : \n",df_movies.isnull().sum(axis=0))
```
